## Copy vs deep copy

```python
import copy

# Create a nested list
original = [1, 2, [3, 4]]

# Shallow copy
shallow_copied = copy.copy(original)

# Deep copy
deep_copied = copy.deepcopy(original)

# Modify the nested list in the original
original[2][0] = 'X'

# Print all lists
print("Original list:", original)
print("Shallow copy:", shallow_copied)
print("Deep copy:", deep_copied)
```

## Network Graphs

```python
import networkx as nx
import matplotlib.pyplot as plt

# Create a figure with 2x2 subplots for the 4 different graphs
fig, axs = plt.subplots(2, 2, figsize=(12, 10))

# 1. A complete graph made of four nodes
G1 = nx.complete_graph(4)
pos1 = nx.circular_layout(G1)
nx.draw(G1, pos1, ax=axs[0, 0], with_labels=True, node_color='lightblue', node_size=500, font_size=12)
axs[0, 0].set_title("Complete Graph (4 nodes)")

# 2. A complete graph made of five nodes
G2 = nx.complete_graph(5)
pos2 = nx.circular_layout(G2)
nx.draw(G2, pos2, ax=axs[0, 1], with_labels=True, node_color='lightgreen', node_size=500, font_size=12)
axs[0, 1].set_title("Complete Graph (5 nodes)")

# 3. A bipartite graph: 2-node group and 3-node group with all possible inter-group edges
G3 = nx.Graph()
# Add nodes with partition attributes
G3.add_nodes_from([0, 1], bipartite=0)  # First group (2 nodes)
G3.add_nodes_from([2, 3, 4], bipartite=1)  # Second group (3 nodes)
# Add all possible edges between groups
for i in range(2):
    for j in range(2, 5):
        G3.add_edge(i, j)
        
pos3 = nx.bipartite_layout(G3, nodes=[0, 1])
nx.draw(G3, pos3, ax=axs[1, 0], with_labels=True, node_color='lightcoral', node_size=500, font_size=12)
axs[1, 0].set_title("Bipartite Graph (2-node and 3-node groups)")

# 4. A bipartite graph: two 3-node groups with all possible inter-group edges
G4 = nx.Graph()
# Add nodes with partition attributes
G4.add_nodes_from([0, 1, 2], bipartite=0)  # First group (3 nodes)
G4.add_nodes_from([3, 4, 5], bipartite=1)  # Second group (3 nodes)
# Add all possible edges between groups
for i in range(3):
    for j in range(3, 6):
        G4.add_edge(i, j)
        
pos4 = nx.bipartite_layout(G4, nodes=[0, 1, 2])
nx.draw(G4, pos4, ax=axs[1, 1], with_labels=True, node_color='lightyellow', node_size=500, font_size=12)
axs[1, 1].set_title("Bipartite Graph (two 3-node groups)")

plt.tight_layout()
plt.show()
```
